```python
import numpy as np

# Create a Python list
my_list = [10, 20, 30, 40, 50]

# Convert list to NumPy array
arr = np.array(my_list)

# Display the array
print("NumPy Array:", arr)

# Print shape of array
print("Shape of array:", arr.shape)

# Print data type of array
print("Data type of array:", arr.dtype)
```

```python
import numpy as np

# Create a 3x4 matrix of zeros
zeros_matrix = np.zeros((3, 4))

# Create a 2x5 matrix of ones
ones_matrix = np.ones((2, 5))

# Print matrices
print("3x4 Zero Matrix:\n", zeros_matrix)
print("Shape:", zeros_matrix.shape)

print("\n2x5 Ones Matrix:\n", ones_matrix)
print("Shape:", ones_matrix.shape)
```

```python
import numpy as np

# Create array from 1 to 12
arr = np.arange(1, 13)

# Reshape into 3x4 matrix
reshaped_arr = arr.reshape(3, 4)

# Display original array
print("Original Array:")
print(arr)
print("Shape:", arr.shape)

# Display reshaped array
print("\nReshaped Array (3x4):")
print(reshaped_arr)
print("Shape:", reshaped_arr.shape)
```

```python
import numpy as np

# Create an array of 10 random integers (between 1 and 100)
arr = np.random.randint(1, 101, 10)

# Display the array
print("Random Array:", arr)

# Find minimum, maximum, mean, and sum
print("Minimum value:", np.min(arr))
print("Maximum value:", np.max(arr))
print("Mean value:", np.mean(arr))
print("Sum of elements:", np.sum(arr))
```

```python
import pandas as pd

# Create a dictionary
data = {
    "Name": ["Tushar", "Anusha", "Parin", "Rahul"],
    "Marks": [85, 92, 78, 88]
}

# Create DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
print("DataFrame:\n")
print(df)

# Print data types of each column
print("\nData Types:\n")
print(df.dtypes)
```
