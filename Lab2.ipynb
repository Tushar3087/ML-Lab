```python
import pandas as pd
import numpy as np

data = pd.read_csv("Iris.csv")

# Create missing value (for demo)
data.iloc[0, 0] = np.nan

# Fill numerical with mean
data['SepalLengthCm'] = data['SepalLengthCm'].fillna(data['SepalLengthCm'].mean())

# Fill categorical with constant
data['Species'] = data['Species'].fillna("Unknown")

print(data.head())
```

```python
import pandas as pd

data = pd.read_csv("Iris.csv")

print("Before:\n", data.dtypes)

# Convert datatype
data['SepalLengthCm'] = data['SepalLengthCm'].astype(float)

print("\nAfter:\n", data.dtypes)
```

```python
import pandas as pd

data = pd.read_csv("Iris.csv")

data = data.rename(columns={
    "SepalLengthCm": "Sepal_Length",
    "SepalWidthCm": "Sepal_Width"
})

print(data.head())
```

```python
import pandas as pd

data = pd.read_csv("Iris.csv")

# Replace incorrect values
data['Species'] = data['Species'].replace("Iris-setosa", "Setosa")

print(data['Species'].unique())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("Iris.csv")

# Line plot
data['SepalLengthCm'].plot(title="Sepal Length Trend")
plt.show()

# Bar plot
data['Species'].value_counts().plot(kind='bar', title="Species Count")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("Iris.csv")

plt.scatter(data['SepalLengthCm'], data['SepalWidthCm'])
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Scatter Plot")
plt.show()
```
