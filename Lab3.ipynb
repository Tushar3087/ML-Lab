```python
import pandas as pd

# Load dataset
data = pd.read_csv("StudentsPerformance.csv")

# Drop Id column if present
if 'Id' in data.columns:
    data = data.drop('Id', axis=1)

# Display first few records
print("First 5 records:\n", data.head())

# Identify numerical & categorical columns
print("\nNumerical columns:\n", data.select_dtypes(include=['int64', 'float64']).columns)
print("\nCategorical columns:\n", data.select_dtypes(include=['object']).columns)

# Select numerical data
num_data = data.select_dtypes(include=['int64', 'float64'])

# Central Tendency
print("\nMean:\n", num_data.mean())
print("\nMedian:\n", num_data.median())
print("\nMode:\n", num_data.mode())

# Dispersion
print("\nMinimum:\n", num_data.min())
print("\nMaximum:\n", num_data.max())
print("\nSum:\n", num_data.sum())
print("\nVariance:\n", num_data.var())
print("\nStandard Deviation:\n", num_data.std())
```

```python
import pandas as pd

data = pd.read_csv("StudentsPerformance.csv")

if 'Id' in data.columns:
    data = data.drop('Id', axis=1)

# Select numerical columns
num_data = data.select_dtypes(include=['int64', 'float64'])

# Quartiles
Q1 = num_data.quantile(0.25)
Q2 = num_data.quantile(0.50)
Q3 = num_data.quantile(0.75)

print("Q1 (25%):\n", Q1)
print("\nQ2 (Median):\n", Q2)
print("\nQ3 (75%):\n", Q3)
```

```python
import pandas as pd

data = pd.read_csv("StudentsPerformance.csv")

if 'Id' in data.columns:
    data = data.drop('Id', axis=1)

# Numerical data
num_data = data.select_dtypes(include=['int64', 'float64'])

# Correlation
print("Correlation Matrix:\n")
print(num_data.corr())

# Covariance
print("\nCovariance Matrix:\n")
print(num_data.cov())
```
